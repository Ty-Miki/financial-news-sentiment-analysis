### Portfolio analysis for  aganist AMZN, GOOG, META, MSFT, NVDA and TSLA

```python
# Portfolio weight analysis of the past 12 years
stocks = ['AAPL', 'AMZN', 'GOOG', 'META', 'MSFT', 'NVDA', 'TSLA']
weights = analyzer.calculate_portfolio_weights(tickers=stocks,
                                               start_date='2012-01-01',
                                               end_date='2024-01-01')
weights
```

```python
# Portfolio performance anaysis of the past 12 years
portfolio_return, portfolio_volatility, sharpe_ratio = analyzer.calculate_portfolio_performance(tickers=stocks,
                                                                                               start_date='2012-01-01',
                                                                                               end_date='2024-01-01')
print(f"Portfolio Return: {portfolio_return:.2f}%")
print(f"Portfolio Volatility: {portfolio_volatility:.2f}%")
print(f"Sharpe Ratio: {sharpe_ratio:.2f}")
```
